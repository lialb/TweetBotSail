```python
import tweepy
import json
import requests
from requests_oauthlib import OAuth1

with open('api_keys.json') as f:
    keys = json.load(f)
```

```python
url = 'https://api.twitter.com/1.1/statuses/user_timeline.json'
params = {
    'screen_name': 'realdonaldtrump',
    'tweet_mode': 'extended',
    'count': '200',
    'include_retweets': 'False',
    'exclude_replies': 'True'
}
auth = OAuth1(keys['key'], keys['secret'], keys['access_token'], keys['access_secret'])
r = requests.get(url, auth=auth, params=params)
```

```python
r.status_code
# print(r.json())
for tweet in r.json():
    print(tweet['full_text'])
```

```python
post_url = 'https://api.twitter.com/1.1/statuses/update.json'
post_params = {
    'status': 'Is requests the best python library :O'
}
p = requests.post(post_url, auth=auth, params=post_params)
```
